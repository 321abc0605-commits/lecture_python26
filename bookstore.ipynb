```python
class Book:

    # 생성자 : 책 객체가 생성될 때 실행
    def __init__(self, title, author, price, stock):

        # 속성(Attribute)
        self.title = title      # 책 제목
        self.author = author    # 저자
        self.price = price      # 가격
        self.stock = stock      # 재고 수량

    # 책 정보 출력 메서드
    def show_info(self):
        print("제목 :", self.title)
        print("저자 :", self.author)
        print("가격 :", self.price)
        print("재고 :", self.stock)

    # 책 판매 메서드
    def sell(self, amount):

        # 재고가 충분한 경우
        if self.stock >= amount:
            self.stock -= amount
            print(amount, "권 판매 완료")

        # 재고 부족
        else:
            print("재고 부족")

    # 재고 추가 메서드
    def add_stock(self, amount):
        self.stock += amount
        print(amount, "권 입고 완료")


# 객체(Object) 생성
book1 = Book("파이썬 입문", "홍길동", 25000, 10)


# 메서드 사용
book1.show_info()

book1.sell(3)

print("남은 재고 :", book1.stock)

book1.add_stock(5)

print("현재 재고 :", book1.stock)
```
