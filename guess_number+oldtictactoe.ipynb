### 숫자맞추기

```python
import random

a = random.randint(1,100)
c = 0

while True:
    b = int(input("맞추기 ㄱㄱ(1-100): "))
    c += 1

    if b > a:
        print("큼")
    elif a > b:
        print("작음")
    else:
        print(f"정답 (시도 횟수: {c})")
        break
```

### 틱택토

```python
import random

a1=a2=a3=b1=b2=b3=c1=c2=c3=" "


print("""
   -입력-
[1] [2] [3]
[4] [5] [6]
[7] [8] [9]
""")

while True:
    print(f"[{a1}] [{a2}] [{a3}]")
    print(f"[{b1}] [{b2}] [{b3}]")
    print(f"[{c1}] [{c2}] [{c3}]")


    n = int(input("입력: "))

    if n == 1 and a1 == " ": a1 = "O"
    elif n == 2 and a2 == " ": a2 = "O"
    elif n == 3 and a3 == " ": a3 = "O"
    elif n == 4 and b1 == " ": b1 = "O"
    elif n == 5 and b2 == " ": b2 = "O"
    elif n == 6 and b3 == " ": b3 = "O"
    elif n == 7 and c1 == " ": c1 = "O"
    elif n == 8 and c2 == " ": c2 = "O"
    elif n == 9 and c3 == " ": c3 = "O"
    else:
        print("잘못 입력")
        continue


    if ((a1==a2==a3=="O") or
        (b1==b2==b3=="O") or
        (c1==c2==c3=="O") or
        (a1==b1==c1=="O") or
        (a2==b2==c2=="O") or
        (a3==b3==c3=="O") or
        (a1==b2==c3=="O") or
        (a3==b2==c1=="O")):
        print(f"[{a1}] [{a2}] [{a3}]")
        print(f"[{b1}] [{b2}] [{b3}]")
        print(f"[{c1}] [{c2}] [{c3}]")
        print("승리")
        break

    while True:
        r = random.randint(1,9)

        if r == 1 and a1 == " ": a1 = "X"; break
        elif r == 2 and a2 == " ": a2 = "X"; break
        elif r == 3 and a3 == " ": a3 = "X"; break
        elif r == 4 and b1 == " ": b1 = "X"; break
        elif r == 5 and b2 == " ": b2 = "X"; break
        elif r == 6 and b3 == " ": b3 = "X"; break
        elif r == 7 and c1 == " ": c1 = "X"; break
        elif r == 8 and c2 == " ": c2 = "X"; break
        elif r == 9 and c3 == " ": c3 = "X"; break

    if ((a1==a2==a3=="X") or
        (b1==b2==b3=="X") or
        (c1==c2==c3=="X") or
        (a1==b1==c1=="X") or
        (a2==b2==c2=="X") or
        (a3==b3==c3=="X") or
        (a1==b2==c3=="X") or
        (a3==b2==c1=="X")):
        print(f"[{a1}] [{a2}] [{a3}]")
        print(f"[{b1}] [{b2}] [{b3}]")
        print(f"[{c1}] [{c2}] [{c3}]")
        print("패배")
        break
```
