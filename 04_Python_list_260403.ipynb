### 리스트의 항목에 접근하기 슬라이싱(slicing)

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]

print(test_list[0:5])
print(test_list[:5])

print(test_list[1::3])
print(test_list[:5:2])
print(test_list[::3])
print(test_list[:])
print(test_list[::])
print('-')
print(test_list[-1:-7:-2])
print(test_list[1:-1:])
```

```python
print([1, 2, 3][::-1])
```

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]
test_list[:2] = [111, 222, 333]
print(test_list)
```

#### 리스트의 항목 재배열하기

```python
a_list = ['one', 'two', 'three', 'four']
print ('원본', a_list)
a_list.reverse()
print('역순 후', a_list)

a_list.sort()
print('정렬 후', a_list)

a_list = ['one', 'two', 'three', 'four']
b_list = sorted(a_list)
print('원본', a_list)
print('정렬 후', b_list)

c_list = sorted(a_list, key=len)
print(c_list)
```

```python
scores = [95, 45, 21, 74, 55]
print(sorted(scores, reverse=True))

print(scores)
scores.sort(reverse=True)
print(scores)
```

#### 리스트에 항목 추가하기

```python
# 1~N의 수로 이루어진 리스트 생성
N = int(input("숫자: "))
num_list = []
for i in range(1,N+1):
      num_list.append(i)
print(num_list)
```

```python
# 1~N의 수로 이루어진 리스트 생성
N = int(input("숫자: "))
num_list = [i for i in range(1,N+1)]
print(num_list)

# expression
num_list = [i * 2 for i in range(1,N+1)]
print(num_list)

# if문 사용
N = int(input("숫자 :"))
num_list = [i for i in range(1,N+1) if i % 2 == 0]
print(num_list)
```

#### 리스트와 반복구조 연습

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def getIndex(A, B):
    index = 0
    while index < len(A):
        if A[index] == B:
            break
        index += 1
    return index

def getMax(A):
    max_val = A[0]
    for B in A:
        if B > max_val:
            max_val = B
    return max_val

def getMin(A):
    min_val = A[0]
    for B in A:
        if B < min_val:
            min_val = B
    return min_val

def countGT(A, B):
    C = 0
    for D in A:
        if D > B:
            C += 1
    return C

def sumList(A):
    B = 0
    for C in A:
        B += C
    return B

def swapList(A):
    B = []
    for C in A:
        B.insert(0, C)
    return B

print(getIndex(number_list, 25))
print(getMax(number_list))
print(getMin(number_list))
print(countGT(number_list, 42))
print(sumList(number_list))
print(swapList(number_list))
print(number_list)
```
