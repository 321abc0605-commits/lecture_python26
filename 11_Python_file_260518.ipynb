```python
fruits = ['사과', '토마토', '바나나']
outfile = 'fruits.txt'
with open(outfile, mode='w', encoding='utf-8') as f_write:
    for fruit in fruits:
        f_write.write(fruit+'\n')
# f_write.writelines(fruits)
# f_write.write('\n'.join(fruits))
```

```python
with open(outfile, mode='a', encoding='utf-8') as f_write:
    f_write.write('\n'.join(fruits))
```

```python
import math
import random

# 반지름 생성
radius = round(random.random() * 10, 2)

# 원의 넓이 계산
area = math.pi * radius ** 2

# 출력
print('원의 반지름:', radius)
print('원주율 pi:', round(math.pi, 4))
print(f'반지름 {radius}인 원의 면적은 {round(area, 2)}')
```

```python
import math
import statistics

if __name__ == '__main__':

    # factorial 출력
    print('1! =', math.factorial(1))
    print('5! =', math.factorial(5))
    print('11! =', math.factorial(11))
    print('16! =', math.factorial(16))

    print()

    # 리스트
    st = [80, 99, 77, 65, 92, 74, 82]

    print(st)

    # 중앙값
    print('중앙값:', statistics.median(st))

    # 평균
    print('평균:', round(statistics.mean(st), 2))

    # 분산
    print('분산:', round(statistics.variance(st), 2))

    # 표준편차
    print('표준편차:', round(statistics.stdev(st), 2))
```
