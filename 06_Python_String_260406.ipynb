### 문자열을 원하는 형식으로: f-문자열

```python
print(f"f-string은 {{변수명}}과 같이 사용된다.")
```

```python
#산술 연산
print(f"2 + 3 = {2 + 3}")

#매서드 호출
lecture = 'python'
print(f"Hello, {lecture.upper()}!")

#조건문
n = 10
print(f"{n}은 {'짝수' if n % 2 == 0 else '홀수'}입니다.")
```

```python
number = 1234.56789

print(f"소수점 자리수 제한: {number:.2f}")

print(f"콤마 추가: {number:,}")

print(f" 동시 사용: {number:,.2f}")
```

## f string


```python
string = 'ABC'
print(f"|{string:10}|")
print(f"|{string:>10}|")
print(F"|{string:^10}|")
print(F"|{string:-^10}|")
print(F"|{string:*>10}|")
```

### format() 메서드, % 서식 지정

```python
# 위치
print("나는 {}와 {}가 좋더라.".format("시골 쥐", "도시 쥐"))

#인덱스 지정
print("{1}이 나중에 {0}으로.".format("구리역", "8호선"))

#이름 지정
print("{과일}는 빨개.".format(과일="사과")) #format 뒤 {}는 안돼더라
```

### 모스 부호

```python
str = input('알파벳 문자열 입력 :')
str = str.upper()

mo = ['·-', '-···', '-·-·', '-··', '·', '··-·', '--·', '····', '··', '·---', '-·-', '·-··', '--', '-·', '---', '·--·', '--·-', '·-·', '···', '-', '··-', '···-', '·--', '-··-', '-·--', '--··']

print('모스부호로 변환 :', end = ' ')
for i in range(len(str)) :
    print(mo[ord(str[i]) - 65], end = ' ')
```
